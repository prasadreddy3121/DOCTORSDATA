# #VISUALIZING DATA

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#missing values
sns.heatmap(dv.isnull(),cbar=False,cmap='viridis')
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(dv.corr(),cbar=True,annot=True,cmap='Blues')
```

```python
#gender graph
plt.figure(figsize=(5,5))
plt.title("GENDER GRAPH",size=15)
plt.xlabel("gender")
plt.ylabel("count")
plt.bar(["female",'male'],dv["gender"].value_counts(), edgecolor ="b", width =0.8)
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.title("freepoor",size=15)
plt.xlabel("govt health insurance")
plt.ylabel("income")
plt.bar(["no","yes"],dv["freepoor"].value_counts(), edgecolor ="b", width =0.8)
plt.show()
dv["freerepat"].value_counts()
```

```python
plt.figure(figsize=(5,5))
plt.title("freerepat ",size=15)
plt.xlabel("govt health insurance")
plt.ylabel("income")
plt.bar(["no","yes"],dv["freerepat"].value_counts(), edgecolor ="b", width =0.8)
plt.show()
dv["freerepat"].value_counts()
```

```python
ds=dv.groupby('gender')['reduced'].sum().to_frame().reset_index()
plt.barh(ds['gender'],ds['reduced'],color=['pink','blue'])
plt.title('bar chart')
plt.xlabel('gender')
plt.ylabel('reduced activity')
plt.show()
```
